### üì¶ Extracting Superstore Dataset from ZIP File

This section handles the extraction of the "Sample - Superstore.csv" dataset from a ZIP file using Python's `zipfile` module and prepares the data directory for analysis.


```python
import zipfile
import os

# Your uploaded ZIP file name
zip_path = "/content/Sample - Superstore.csv.zip"

# Create a folder to extract files
extract_folder = "/content/superstore_data"

# Extract ZIP contents
with zipfile.ZipFile(zip_path, 'r') as zip_ref:
    zip_ref.extractall(extract_folder)

# List extracted files
os.listdir(extract_folder)
```

### üìÑ Loading Superstore Dataset with Proper Encoding

This section loads the extracted `Sample - Superstore.csv` file using `pandas`, ensuring correct encoding (`ISO-8859-1`) to handle special characters. It then displays the first 5 rows of the dataset for initial inspection.


```python
import pandas as pd

# Load CSV with correct encoding
df = pd.read_csv("/content/superstore_data/Sample - Superstore.csv", encoding='ISO-8859-1')

# Preview first 5 rows
df.head()
```

### üßπ Removing Duplicate Records

To ensure data quality, we remove any duplicate rows from the dataset:

```python
# Remove duplicate rows
df = df.drop_duplicates()
print("Duplicates removed. Current shape:", df.shape)
```

### üí∞ Key Business Metrics Summary

Prints the total sales, total profit, and total quantity sold across the dataset to get a high-level view of performance:


```python
print("Total Sales: $", round(df['Sales'].sum(), 2))
print("Total Profit: $", round(df['Profit'].sum(), 2))
print("Total Quantity Sold:", df['Quantity'].sum())
```

### üåç Sales and Profit by Region

Groups the dataset by **Region** and calculates the **total Sales and Profit** for each region. The results are sorted in descending order of sales to identify the highest-performing regions.


```python
region_sales = df.groupby('Region')[['Sales', 'Profit']].sum().sort_values(by='Sales', ascending=False)
region_sales
```

### üèÜ Top 10 Customers by Sales

Identifies the top 10 customers based on their **total sales contribution**. This helps in recognizing the most valuable clients for the business.


```python
top_customers = df.groupby('Customer Name')['Sales'].sum().sort_values(ascending=False).head(10)
top_customers
```

### üìä Sales & Profit by Category and Sub-Category

This analysis groups the data by **Category** and **Sub-Category**, calculating the total **Sales** and **Profit** for each combination. The top 10 entries (by sales) are shown to highlight the most revenue-generating product segments.


```python
category_sales = df.groupby(['Category', 'Sub-Category'])[['Sales', 'Profit']].sum().sort_values(by='Sales', ascending=False)
category_sales.head(10)
```

### üìà Monthly Sales Trend Analysis

This section converts the `Order Date` column to datetime format and then groups sales data by **month** to analyze trends over time. The resulting line chart visualizes how sales fluctuate across months.


```python
# Convert to datetime
df['Order Date'] = pd.to_datetime(df['Order Date'])

# Monthly sales
monthly_sales = df.groupby(df['Order Date'].dt.to_period('M'))['Sales'].sum()
monthly_sales.plot(kind='line', figsize=(10,5), title='Monthly Sales Trend')
```

### üé® Setting Up Visualization Environment

Before plotting any charts, we import the required

```python
import matplotlib.pyplot as plt
import seaborn as sns

# Set style
sns.set(style="whitegrid")
```

### üìä Sales by Region (Horizontal Bar Chart)

This visualization shows the total sales across different regions using a horizontal bar chart. The data is grouped by `Region`

```python
region_sales = df.groupby('Region')['Sales'].sum().sort_values()
region_sales.plot(kind='barh', color='skyblue', figsize=(8,5), title='Sales by Region')
plt.xlabel('Sales')
plt.ylabel('Region')
plt.show()
```

### üí∞ Profit by Product Category (Bar Chart)

This visualization displays the total profit generated by each product category. The chart helps in identifying which categories are more profitable based on the aggregated profit values.

```python
category_profit = df.groupby('Category')['Profit'].sum()
category_profit.plot(kind='bar', color='lightgreen', figsize=(6,4), title='Profit by Category')
plt.ylabel('Profit')
plt.show()
```

### üìä Top 10 Sub-Categories by Sales (Horizontal Bar Chart)

This bar chart highlights the top 10 sub-categories with the highest total sales. It helps identify which product sub-categories are the most revenue-generating, offering key insights for inventory planning and sales strategies.


```python
top_subcat = df.groupby('Sub-Category')['Sales'].sum().sort_values(ascending=False).head(10)
sns.barplot(x=top_subcat.values, y=top_subcat.index, palette='mako')
plt.title('Top 10 Sub-Categories by Sales')
plt.xlabel('Sales')
plt.show()
```


### üìà Sales vs. Profit by Category (Scatter Plot)

This scatter plot visualizes the relationship between sales and profit across different product categories. It helps in identifying high-sales but low-profit segments, or categories that consistently yield higher profitability.


```python
plt.figure(figsize=(8,5))
sns.scatterplot(x='Sales', y='Profit', hue='Category', data=df)
plt.title('Sales vs Profit by Category')
plt.show()
```

### üöö Sales by Shipping Mode

This horizontal bar chart shows total sales grouped by different shipping modes. It provides insights into which shipping options are most popular or contribute the most to revenue.


```python
ship_sales = df.groupby('Ship Mode')['Sales'].sum().sort_values()
ship_sales.plot(kind='barh', color='orange', figsize=(7,4), title='Sales by Ship Mode')
plt.xlabel('Sales')
plt.ylabel('Ship Mode')
plt.show()
```

### üßë‚Äçüíº Sales and Profit by Customer Segment

This grouped bar chart displays the total Sales and Profit across different customer segments (Consumer, Corporate, Home Office). It highlights how each segment contributes to the overall business revenue and profitability.


```python
segment_data = df.groupby('Segment')[['Sales', 'Profit']].sum().sort_values(by='Sales', ascending=False)
segment_data.plot(kind='bar', figsize=(7,5), color=['#007acc', '#00cc99'], title='Sales and Profit by Segment')
plt.ylabel('Amount')
plt.show()
```

### üó∫Ô∏è Top 10 States by Sales

This horizontal bar chart highlights the top 10 U.S. states with the highest total sales. It provides a quick snapshot of regional performance, helping identify key markets contributing the most to overall revenue.


```python
top_states = df.groupby('State')['Sales'].sum().sort_values(ascending=False).head(10)
sns.barplot(x=top_states.values, y=top_states.index, palette='coolwarm')
plt.title('Top 10 States by Sales')
plt.xlabel('Sales')
plt.show()
```

### üìä Correlation Matrix of Key Metrics

The heatmap below visualizes the correlation between key numerical features: **Sales**, **Quantity**, **Discount**, and **Profit**

```python
plt.figure(figsize=(8,6))
sns.heatmap(df[['Sales', 'Quantity', 'Discount', 'Profit']].corr(), annot=True, cmap='Blues')
plt.title('Correlation Matrix')
plt.show()
```

### üí∏ Discount vs Profit by Category

This scatter plot illustrates the relationship between **Discount** and **Profit** across different product categories. Each point represents an order, colored by category.


```python
plt.figure(figsize=(7,4))
sns.scatterplot(data=df, x='Discount', y='Profit', hue='Category')
plt.title('Discount vs Profit by Category')
plt.show()
```

### ü•ß Sales Share by Category

This pie chart visualizes the proportion of total **Sales** contributed by each product **Category**. It highlights which segments dominate the sales distribution.



```python
category_share = df.groupby('Category')['Sales'].sum()
category_share.plot(kind='pie', autopct='%1.1f%%', startangle=140, figsize=(6,6), title='Sales Share by Category')
plt.ylabel('')
plt.show()
```

After performing all data cleaning, transformation, and feature engineering steps, we save the final cleaned dataset for further analysis or dashboard development.
